### Hyperparameters
- In statistics, hyperparameter is a parameter from a prior distribution; it captures the prior belief before data is observed. In any machine learning algorithm, these parameters need to be initialized before training a model.
- A hyperparameter is a parameter whose value is set before the learning process begins.
- Hyperparameters are important because they directly control the behaviour of the training algorithm and have a significant impact on the performance of the model is being trained.
- Example of algorithm hyperparameters can be `learning rate`, `k in k-Nearest Neighbors`, `the number of hidden layers in Neural Networks`, etc.

### Hyperparameter tuning
- Hyperparameter tuning(or hyperparameter optimisation) is choosing a set of optimal hyperparameters for a learning algorithm.
- Some basic algorithms for this are:
    * Grid Search
    * Random Search

### Grid search
- It is a traditional way to perform hyperparameter optimization. It works by searching exhaustively through a specified subset of hyperparameters.

### Random Search
- Random search differs from grid search mainly in that it searches the specified subset of hyperparameters randomly instead of exhaustively. The major benefit being decreased processing time.
- There is a tradeoff to decreased processing time, however. We arenâ€™t guaranteed to find the optimal combination of hyperparameters.

```python
# Import necessary packages
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.model_selection import GridSearchCV
```

```python
df = pd.read_csv("/home/triloq/stuff/projects/machine_learning_algorithms/extra/data/breast_cancer_dataset.csv")
```

```python
df
# Each row in the dataset have one of two possible diagnosis: benign (B) and malignant (M)
```

```python
# Pre-process
# Rename diagnosis column value, making B to 0 and M to 1
df["diagnosis"] = np.where(df["diagnosis"]=='M', 1, 0) # If value if M then make it 1 else make it 0
# Remove NaN value column
df = df.drop(["Unnamed: 32", "id"], axis=1)
df
```

```python
df["diagnosis"].value_counts()
```

```python
# Splitting the dataset
X_1 = df.drop(['diagnosis'], axis=1)
y = df['diagnosis']
# Normalizing X values before splitting(using min-max normalization)
# If we won't normalize here, then we will get all values as 0 for y_pred
X = (X_1 -np.min(X_1))/(np.max(X_1)-np.min(X_1)).values
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
# Here we are using Logistic Regression for now
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
y_pred = logreg.predict(X_test)
y_pred
```

```python
print('Accuracy Score : ' + str(accuracy_score(y_test, y_pred)))
print('Precision Score : ' + str(precision_score(y_test, y_pred)))
print('Recall Score : ' + str(recall_score(y_test, y_pred)))
print('F1 Score : ' + str(f1_score(y_test, y_pred)))
```

```python
print('Confusion Matrix : \n' + str(confusion_matrix(y_test,y_pred)))
```

We can see that there are 6 malignant cases have been classified incorrectly as benign (False negatives) and 0 benign case has been classified as malignant (False positive).
A false negative is more serious as a disease has been ignored, which can lead to the death of the patient. At the same time, a false positive would lead to an unnecessary treatment â€” incurring additional cost.

Letâ€™s try to minimize the false negatives by using Grid Search to find the optimal parameters. Grid search can be used to improve any specific evaluation metric.
The metric we need to focus on to reduce false negatives is Recall.


The grid search provided by GridSearchCV exhaustively generates candidates from a grid of parameter values specified with the param_grid parameter. For instance, the following param_grid:
Some common parameters for Logistic Regression:

- C: Inverse of regularization strength- smaller values of C specify stronger regularization.
- Penalty: l1 or l2 which species the norm used in the penalization.
- Kernel: We can set the kernel parameter to linear, poly, rbf, sigmoid, precomputed or provide our own callable.
- Gamma: This is the coefficient for rbf, poly and sigmoid kernel parameter.

**For penalty parameter: [Read this](https://www.statisticshowto.com/regularization/)**

**For C parameter: [Read this](https://www.quora.com/What-is-the-C-parameter-in-logistic-regression)**

```python
log_reg = LogisticRegression()
grid_values = {'penalty': ['l1', 'l2'],'C':[0.001,.009,0.01,.09,1,5,10,25]}
# GridSearchCV library does a exhaustive search over specified parameter values for an estimator.
grid_clf = GridSearchCV(log_reg, param_grid = grid_values,scoring = 'recall')
```

```python
grid_clf.fit(X_train, y_train)
y_pred_new = grid_clf.predict(X_test)
y_pred_new
```

```python
# New Model Evaluation metrics 
print('Accuracy Score : ' + str(accuracy_score(y_test, y_pred_new)))
print('Precision Score : ' + str(precision_score(y_test, y_pred_new)))
print('Recall Score : ' + str(recall_score(y_test, y_pred_new)))
print('F1 Score : ' + str(f1_score(y_test, y_pred_new)))
confusion_matrix(y_test,y_pred_new)
```
