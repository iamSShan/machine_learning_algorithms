## Introduction:
* Our life cycle of using ML algoithm is:
    * Data preprocessing
        * 1) EDA
        * 2) Feature Engineering
        * 3) Feature Selection
        
    * Model training and testing
    * Model deployment
        
* In scikit-learn for data pre-processing we have object called Tranformers. e.g: StandardScaler, MinMaxScaler, PCA, Imputer(for handling missing values) etc.
* For model creation, in scikit learn we have multple models like Linear Regression, Random Forest, etc.
* **Transformers** just change input data into some other format i.e. transform data and then we use this transformed data in model training.

### With respect to Data preprocessing:

#### fit and transform:

Now let's first talk in terms of Transformers, let's say we are taking about StandardScaler

Let's suppose we have some features like f1, f2, f3, etc. So in this standardization process, we let's say want to standardize f1 so it will convert a feature f1 to f1’ for each row using above formula.

* This here we are converting one input feature and converting it. This is transformation. In scikit-learn, we use fit and transform to do this.
    * **Inside fit method**- It just computes μ and σ values (in case of standard scaler)
    * **Inside transform**- We apply this entire formula to each row value of f1
    
* `transform` is only applied when we want to tranform the input data.
* This transformed data is given to model as input.    

#### fit_transform
* It just performs both the step fit and transform all together

* In case of transformers, `Fit` and `transform` (*fit_transform*) is used for training data and for test data we only have to do `transform`.

* Interview question: Why `Fit` and `transform` for train and only `transform` for test?
    * Because fit only calculates terms of formula and we calculate it once and we use same formula for testing data, so we don't have to calculate two times. It helps us to overcome overfitting.



### With respect to Model creation:
* Here we only do two steps:
    * fit
    * predict
* Let's say after transforming we have features like f1’, f2’, f3’ and we give it to our model then:
    * **In fit method**: It will learn from this input data and will setup it's parameters and weights. For example in linear regression; y=mx+c. It will try to learn coefficient values m and c based on this data. When it learns then model is ready.
    * **In predict method**: When new data comes, and we pass it our model, this model will do prediction and give us output.
    
* `Fit` is used for training data and `predict` is used for testing data.
