## Auto Regression(AR Models):
* Regression: Used to predict continuous value of an item based on certain parameters
* Auto: Uses its own past values to predict future values
    
      yₜ = C₀ + C₁yₜ₋₁ // AR(1): 1st order auto regression
      yₜ = C₀ + C₁yₜ₋₁ + C₂yₜ₋₂ // AR(2): 2nd order auto regression
      // C0, C1, C2... are constants
* For eg: temperature can depend on previous days
* Example 2: Let's say we have marks of student throughout school from 7th to 11th standard and we want to predict marks in 12th standard
    * 7th: 59, 8th: 60, 9th: 72, 10th: 92, 11th: 88
    * So we can use first order regression(11th marks) and may be 10th marks also helps in determining so we can use second order too:
    * Therfore, **12th marks: C0 + C1* (11th marks) + C2 * 10th marks + Error**
    * Now it may be case that 8th class marks are also related how the marks are scored 4 years later. We need to identify this relation, for this we use:

* **Auto-Correlation**: Correlation is relationship b/w 2 variables. And Auto Correlation is relationship of a variable with its previous time period values(lags). To quantify this we use Pearson's Correlation Coefficient(values b/w -1 and 1).
* In auto correlation there be 2 functions:
    * **Auto-Correlation Function** : There can be direct and indirect effect of values. For eg: 8th class can be directly related to class 12th or it can be related to class 9th and class 9th then is related to class 12th.
    * **Partial Auto Correlation**: Only direct effect of values in previous time lags.
    
* In case of Auto Regression we use **Partial Auto Correlation**, because we want to check whether directly 12th class marks are related to 8th class or not.
* We can plot the correlation values for each lag and check till when the correlation value is good.


## Moving Average Model:
* It is a model that predicts future values of a time series using the past errors.
        
        AR(1) auto regression : yₜ = C₀ + C₁yₜ₋₁ + Єₜ
        MA(1) moving average: yₜ = C₀ + θ₁Єₜ₋₁ + Єₜ
       // Єₜ means error current time, Єₜ₋₁ means error in prev time and θ is constant
* Here we use Auto correlation function, where in Auto regression we use partial auto correlation function.
        
## ARMA Model(Auto Regression Moving Average Model):
* A simple combination of auto regression and moving average model.
* So eqn here looks like: yₜ = C₀ + C₁yₜ₋₁ + θ₁Єₜ₋₁ + Єₜ -----> *ARMA(1,1)*
* Above eqn is represented as ARMA(1,1), it means first order auto regression and 1st order moving average. We can also calculate for different orders too.
* **How to find the order?** Again we use ACF(auto correlation function) and PACF(partial auto correlation function). So here we plot both the models.

* 

    * On the left we have PACF and on right we have ACF, on x-axis we have time periods(1 means 1 time period back) and on y-axis we have correlation.
    * In PACF we can see we have high values of correlation, which means 1st and 2nd time period lags are relevant
    * In ACF we can see we get significant values for 1, 2 and 3 lags which makes it as 3rd order moving average model
    * Therefore ARIMA(2,3)

## ARIMA Model(Auto Regression Integrated Moving Average Model):
* It is almost ARMA, just one simple operation we do first of converting non stationary time series to stationary time series and do then do same as ARMA.
* We use a differencing operation to convert non stationary series to stationary series. Differencing notes in first notebook.
* Here the notation is ARIMA(P, D, Q)
    * P: Order of AR model
    * D: Order of differencing to get stationary series
    * Q: Order of MA model

## Vector Auto Regression Model:
* It is slight modification of Auto Regression model.
* What AR model does: Calculates future values of a variable based on its previous time lags. While VAR model calculates future values of a variable based on its previous time lags and the previous lags of some other dependent variable.
* Let's say we are calculating car price, wrt AR it will be dependent on car price value yesterday and prev days.
* But in VAR, car price might also depend on other variables like cost of public transport, etc.
* If we assume, if public transport cost increases, car price also increases, if it decreases then car price also decreases. Let's say public transport Price X, then VAR(1) model:

        yₜ = C₀ + C₁₁yₜ₋₁ + C₁₂Xₜ₋₁+ Єₜ₁
        Xₜ = C₀ + C₂₁Xₜ₋₁ + C₂₁yₜ₋₁+ Єₜ₂
