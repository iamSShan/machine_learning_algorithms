* **Time series problem**: Let's say if on specific day `t`, sales of Ac was something, then if we want to predict for `t+1` what are the sales. And here we are dependent on the previous day `t` data. And this is called as extrapolation. So we can say in case of time series for `t` day sales we are dependent on t-1 , t-2, t-3, etc, which is also called as lags(one time stamp lag, 2 timestamp lag, etc)

* Whether we wish to predict the trend in financial markets or electricity consumption, time is an important factor that must now be considered in our models. For example, it would be interesting to forecast at what hour during the day is there going to be a peak consumption in electricity, such as to adjust the price or the production of electricity. We can use time series analysis.

* Time series analysis is a specific way of analyzing a sequence of data points collected over an interval of time. In time series analysis, analysts record data points at consistent intervals over a set period of time rather than just recording the data points intermittently or randomly.
* In a time series, time is often the independent variable and the goal is usually to make a forecast for the future.
* However, there are other aspects that come into play when dealing with time series.
 
    * Stationarity: A time series is said to be stationary if its statistical properties do not change over time. In other words, it has constant mean and variance, and covariance is independent of time. Also no seasonality. We want out model to be stationary in time series
    
    
    * Is there Seasonality?: It means there are repeating trends/patterns over time.
    
    * Is the target variable autocorrelated?(Autocorrelation is the similarity between observations as a function of the time lag between them.)

### How can we convert non stationary to stationary
* Some methods are:

* **Differencing**: Here we subtract the values of time series with the value at some different time period. First order differencing is subtracting with values 1 time period back, we can also do second order differencing where we subtract with values with 2 time period back. Below is an image for first order differencing.



* Log operations to smooth exponential curve, as Log(exp(x)) = x
* Seasonal differencing: Here instead of subtracting with just previous, we subtract with time period which is repeating. Y(t) - Y(t-N); where N is time cycle where pattern is repeating.

### White Noise:
* When we are doing any time series modeling project and want to make future predictions, the moment we realize that a particular series is white noise, there we abandon the project. Because then it is not possible to forecast the values for future.
* Characteristics of white noise:
    * Constant Mean and Constant Variance
    * And No auto correlation: Means value at particular period has no relation with values at past time period.
