## Covariance:

Covariance is a measure of how two random variables change together. It tells us whether an increase in one variable corresponds to an increase or decrease in another variable.

Key Characteristics:
* Positive Covariance: If the variables tend to increase together, the covariance is positive.
* Negative Covariance: If one variable tends to increase when the other decreases, the covariance is negative.
* Zero Covariance: If the variables are independent of each other, the covariance is close to zero.

Formula:
For two variables X and Y with means μX and μY:
    
    Cov(X, Y) = ∑(Xi−μX)(Yi−μY)/n−1
Where n is the number of data points.

Example:
Suppose we have the following pairs of values for XX and YY:

    X=[1,2,3,4,5] and Y=[2,4,6,8,10]

The covariance of X and Y would be positive, indicating that as X increases, Y also increases.

**Disadvantage**: Covariance cannot describe magnitude of correlation.

## Pearson Correlation:


Pearson correlation (also known as Pearson's r) measures the linear relationship between two variables. **It gives both the direction and the strength of the relationship.**

Key Characteristics:

* Range: The Pearson correlation coefficient ranges from -1 to 1.
  * +1: Perfect positive linear relationship.
  * 0: No linear relationship.
  * -1: Perfect negative linear relationship.

* Linear Relationship: Only captures linear relationships between variables.

Formula: For two variables X and Y:
    
    r = Cov(X,Y)/σXσY

    Where σX and σY are the standard deviations of X and Y, respectively.

Example:
Using the same values as before:

    X=[1,2,3,4,5] and Y=[2,4,6,8,10]

The Pearson correlation would be +1, indicating a perfect positive linear relationship.


## Spearman Correlation

Spearman correlation measures the strength and direction of the monotonic relationship between two variables. It assesses how well the relationship between two variables can be described using a monotonic function. 

**Monotonic Relationship**: Captures both linear and non-linear monotonic relationships.

Key Characteristics:

* Range: The Spearman correlation coefficient ranges from -1 to 1.
  * +1: Perfect positive monotonic relationship.
  * 0: No monotonic relationship.
  * -1: Perfect negative monotonic relationship.
    

Formula:
For two variables X and Y:

1. Rank the data points in XX and YY.
2. Calculate the Pearson correlation coefficient on the ranked variables R(X)and R(Y).

Example:
Using the same values as before:

    X=[1,2,3,4,5] and Y=[2,4,6,8,10]

The Spearman correlation would also be +1, indicating a perfect positive monotonic relationship.


### Summary:

* Covariance: Measures how two variables change together. Positive covariance means they increase together, negative means one increases while the other decreases.
* Pearson Correlation: Measures the linear relationship between two variables, ranging from -1 to 1. A value close to +1 or -1 indicates a strong linear relationship, while a value close to 0 indicates no linear relationship.
* Spearman Correlation: Measures the monotonic relationship (not necessarily linear) between two variables, also ranging from -1 to 1. It is based on the ranked values of the variables.
    
**Note**:
Monotonic relationships can indeed be non-linear, but not all non-linear relationships are monotonic. Here's a more detailed explanation:


A monotonic relationship between two variables is one in which the variables move in a consistent direction relative to each other. Specifically:
* Monotonically Increasing: As one variable increases, the other variable either always increases or stays the same.
* Monotonically Decreasing: As one variable increases, the other variable either always decreases or stays the same
