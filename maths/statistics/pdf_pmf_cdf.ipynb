### What are Random Variables:
* A random variable is a variable in statistics and probability theory that can take on different values as a result of random events or processes. It represents the outcomes of a random phenomenon, where each outcome has a certain probability associated with it.

* In simple words: **A random variable is a set of possible values from random experiment.** Example of random experiment can be: coin toss, rolling of dice.
* In coin toss random variable: X = {1,0} ; In dice roll Y = {1,2,3,4,5,6}
* Random variables are usually denoted by capital letters, while normal algebra variables are denoted by small letters.
* Think of a random variable like a box that holds different numbers, but we don't know which number we're going to pick until we reach inside and grab one.

* In math, we use random variables to describe these uncertain situations. They help us understand the chances of different outcomes happening and make predictions about what might happen next in all kinds of situations, like games, experiments, or real-life events. So, a random variable is just a way of talking about the different possibilities that can happen randomly.

* **Types of random variable**:
    * Discrete random variable
        * Coin toss(H, T)
        * Dice roll(1, 2, 3, 4, 5, 6)
    * Continuous random variable
        * CGPA of students of a given school class(here it will be between 0 to 10, but it can be anything like 8.7, 9.9, etc.



### Probability Distribution Function:
* A probability distribution is a list of all of the possible outcomes of a random variable along with their corresponding probability values.

**Let's understand with help of example**:

* Imagine we playing a game where you roll a fair six-sided die. Each side has an equal chance of landing face-up. Now, if you wanted to know the chances of rolling a specific number, say 3, you'd say it's 1 out of 6, because there's only one 3 on the die out of six total sides.
* Now, let's think about all the possible outcomes of rolling the die and their probabilities. A probability distribution function (PDF) tells us these probabilities for every possible outcome. So, for our six-sided die example:

    *  The PDF would say that rolling a 1 has a probability of 1/6 (because there's only one 1 on the die).
    * Rolling a 2, 3, 4, 5, or 6 each also has a probability of 1/6 because there's one of each on the die, out of six total sides.

* The important thing about the PDF is that it gives us the likelihood of each outcome happening. It helps us understand the randomness of events and make predictions based on those probabilities. So, in a nutshell, the PDF tells us the chances of different outcomes occurring for a random event.

And there is one more distribution function, which can formed using PDF or PMF: 
* **CDF**: Cummulative Distribution Function

### Cummulative Distribution Function:
* The CDF gives us the probability that a random variable takes on a value less than or equal to a given value.
* It applies to both discrete and continuous random variables.
* For discrete random variables, the CDF is found by summing up the probabilities from the PMF up to a certain point.
* For continuous random variables, the CDF is found by integrating the PDF up to a certain point.
* For example, if you have the CDF for the roll of a fair six-sided die, it would tell you the probability of rolling a number less than or equal to a certain value. So, for rolling a 3 or less on a fair six-sided die, the CDF would give you 3/6 = 1/2.

#### How is PDF graph formed?
* Ans: Using Density Estimation
* Density estimation is a statistical technique used to estimate the probability density function (PDF) of a random variable based on a set of observations or data. In simpler terms, it involves estimating the underlying distribution of a set of data points.
* Density estimation can be used for a variety of purposes, such as hypothesis testing, data analysis, and data visualization. It is particularly useful in areas such as machine learning, where it is often used to estimate the probability distribution of input data or to model the likelihood of certain events or outcomes.
* There are various methods for density estimation, including **parametric and non-parametric approaches**. 

* **Parametric methods assume that the data follows a specific probability distribution (such as a normal distribution) and we do task according to that distribution, while non-parametric methods do not make any assumptions about the distribution(or when it is not matching with any kind of distribution) and instead estimate it directly from the data**. Commonly used techniques for density estimation include kernel density estimation (KDE), histogram estimation, and Gaussian mixture models (GMMs).

* The choice of method depends on the specific characteristics of the data and the intended use of the density estimate.

## Probability Mass Function:
* The PMF is used to describe the probabilities(probability distribution) of discrete random variables.
* Imagine we have a discrete random variable, like the outcome of rolling a fair six-sided die. The PMF tells we the probability of each possible outcome.
* The probabilities assigned by the PMF must satisfy two conditions:
  1. The probability assigned to each value must be non-negative (i.e., greater than or equal to zero).
  2. The sum of the probabilities assigned to all possible values must equal 1.

* For example, if we roll a fair six-sided die, the PMF would tell us that the probability of rolling a 1 is 1/6, the probability of rolling a 2 is also 1/6, and so on.
* So PMF here(mathematical equation in terms y=f(x) will be:

        y = { 1/6 if x  {1, 2, 3, 4, 5, 6}
              0 otherwise}
* If we draw graph, **in PMF: y-axis represents probability and x-axis represents values.** In PDF y-axis was reprensenting probability density.

* Eg of PMF distribution:
  * Binomial distribution
  * Poisson distribution

* In a dice-rolling example, the probability of getting one outcome would be 1/6(0.16) since the outcomes would be independent. Here probability distribution will look like:





```python
import random
import pandas as pd

# Let's say we roll dice 10000 times
L = []
for i in range(10000):
    a = random.randint(1,6)
#     b = random.randint(1,6)

    L.append(a)
print(L[:5])

# Now we can plot graph to see how our PMF looks like in this case:

# First we convert out list to series and we will divide by total number of value we have
# This gives probability values for each of the outcomes
s = (pd.Series(L).value_counts()/pd.Series(L).value_counts().sum()).sort_index()
# We see all probabilities are near 0.16(which is 1/6 for each outcome)
s
```

```python
# To plot:
s.plot(kind='bar')
# Got the same barchart like above image
# Theoritically it should have been 0.16 for all but as it is random experiment hence it is slightly different
```

**Cumulative Distribution Function(CDF) of PMF**:
* The cumulative distribution function (CDF) F(x) describes the ***probability that a random variable X with a given probability distribution will be found at a value less than or equal to x***

* Let X be a discrete random variable with a PMF p(x). The CDF of X, denoted by F(x), is defined as follows: 

        F(x) = P(X  x) = p(i) for all i  x

* So basically CDF of PMF gives what will be probability less than given value. Like probability of X<=4 (dice roll gives less than equal to 4)

* Given our simple Example. For a dice roll, the possible outcomes for X(discrete random variable) are 1,2,3,4,5 or 6. The PMF of X is given by:

        p(1) = 1/6, p(2) = 1/6, p(3) = 1/6, p(4) = 1/6, p(5) = 1/6, p(6) = 1/6

* To find the CDF of X, we can add up the probabilities of the outcomes up to and including each value of X:

        F(1) = P(X  1) = p(1) = 1/6
        F(2) = P(X  2) = p(1) + p(2) = 1/6 + 1/6 = 1/3
        F(3) = P(X  3) = p(1) + p(2) + p(3) = 1/6 + 1/6 + 1/6 = 1/2
        F(4) = P(X  4) = p(1) + p(2) + p(3) + p(4) = 1/6 + 1/6 + 1/6 + 1/6 = 2/3
        F(5) = P(X  5) = p(1) + p(2) + p(3) + p(4) + p(5) = 1/6 + 1/6 + 1/6 + 1/6 + 1/6 = 5/6
        F(6) = P(X  6) = p(1) + p(2) + p(3) + p(4) + p(5) + p(6) = 1/6 + 1/6 + 1/6 + 1/6 + 1/6 + 1/6 = 1

* This means that the probability of getting 1 or fewer on a dice roll is 1/6, the probability of getting 3 or fewer is 1/2, and the probability of getting 6 or fewer is 1.   

* **Summarizing: PMF gives us probability of each particular point, while CDF gives probability of all the points up to that value.**
