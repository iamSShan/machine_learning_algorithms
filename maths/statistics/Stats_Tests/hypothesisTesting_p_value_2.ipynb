## P-value:
* P-value is the **probability** of getting a sample as or more extreme(having more evidence against H0) than our own sample given the Null Hypothesis(H0) is true.
* In simple words p-value is a measure of the strength of the evidence against the Null Hypothesis that is provided by our sample data.

* Let's say we are considering a experiment where we have a coin and we toss it 100 times. And we are noting down no. of heads.
* This coin is fair, so null hypothesis can be P(H) = P(T)
* But there is researcher who says other hypothesis that P(H) > P(T), which is our alternate hypothesis.
* The probability will be mostly around 50 as much time we conduct the whole experiment



* Now let's say we did the experiment and we got 53 heads, so based on just this single experiment can we say that any of the hypothesis is True? Answer is No. Hence we need to do hypothesis testing.
* Now what will be our P-value according to above definition: Prob of head coming greater than 53. We are already at 53, if we are getting more than 53, which means heads are always more, as definition is probability of getting sample(more values than 53) as or more extreme than our own sample(53).


* So in above diagram, if we take 53 from x-axis and take area of curve on right side, that whole area is p-value. As show below:



* **Let's say our p-value here is 0.3, which means if we conduct this experiment 100 times(100 times tossing coins 100 times) then it 0.3 means 30% of time i.e 30 out of 100 times 53 heads can come**.


* Let's consider a case where we get 80 Heads, here we get p-value as 0.00 as area of curve is 0 in this case



* So here we can say if null hypothesis is true, and we perform experiment 100 times then probability of heads coming 80 times is 0. And if by chance 80 heads are coming which means null hypothesis is not true(coin in not fair).


**Now in which scenario we reject null hypothesis**:
* We take a (significance value) which is usually 0.05 or 0.01
* A general thumb rule is if our **p-value <=  then we reject null hypothesis**.

* Let's say we got 80 heads, we get p-value as 0.00, here p-value <  (0.05) so we reject the null hypothesis, which means we say coin is not fair.
* Let's say in case of 60 heads we get p-value as 0.02, here p-value <  (0.05) so we reject the null hypothesis, which means we say coin is not fair.
* Let's say in case of 53 heads we get p-value as 0.30, here p-value > (0.05) so we failed to reject null hypothesis. There is not enough evidence to conclude that coin is not fair.



In some scenarios we won't have significance value() then in that case:
1. Very small **p-values (e.g., p < 0.01) indicate strong evidence** against the null hypothesis, suggesting that the observed effect or difference is unlikely to have occurred by chance alone.

2. Small **p-values (e.g., 0.01  p < 0.05) indicate moderate evidence** against the null hypothesis, suggesting that the observed effect or difference is less likely to have occurred by chance alone.

3. Large **p-values (e.g., 0.05  p < 0.1) indicate weak evidence** against the null hypothesis, suggesting that the observed effect or difference might have occurred by chance alone, but there is still some level of uncertainty.

4. Very large **p-values (e.g., p  0.1) indicate weak or no evidence** against the null hypothesis, suggesting that the observed effect or difference is likely to have occurred by chance alone.




## P-value in context of Z-test:

### Example 1:
Suppose a company is evaluating the impact of a new training program on the productivity of its employees. The company has data on the average productivity of its employees before  implementing the training program. The average productivity was 50 units per day. After implementing the training program, the company measures the productivity of a random sample of 30 employees. The sample has an average productivity of 53 units per day and the pop std is 4. The company wants to know if the new training program has significantly increased productivity.

**Solution**:
    
    We are give population mean() = 50 
    n = 30   
    sample mean(x) = 53
    population std dev() = 4


Now we need to do hypothesis testing

* H0 = mean is still 50 after training program
* Ha = mean is > 50 // as significantly increased is written in question so >

* Let's take significance value () = 0.05

* Then we will calculate Z-statistic:

        (x-)/ (/sqrt(n)) = 4.10

* Then we draw graph and mark 4.10 in the graph, now we find area of the curve on the right of this point, as we need to find probability of current sample value or more extreme than it.



* Then from Z-table we need to find area for this point, but Z-table will give left side area, so we will do 1 - area
* Let's say 0.999 is area, then our rhs area = 1-0.999 = 0.001
* This is our p-value: 0.001

* Now we will compare p-value with  value and if p-value <  then we reject null hypothesis
* Therefore 0.001 < 0.05, hence we can say that we can reject null hypothesis and productivity has increased after training program

### Example 2:

Suppose a snack food company claims that their Lays wafer packets contain an average weight of 50 grams per packet. To verify this claim, a consumer watchdog organization decides to test a random sample of Lays wafer packets. The organization wants to determine whether the actual average weight differs significantly from the claimed 50 grams. The organization collects a random sample of 40 Lays wafer packets and measures their weights. They find that the sample has an average weight of 49 grams, with a pop standard deviation of 5 grams.

**Solution**:

    We are give population mean() = 50 gms
    n = 40   
    sample mean(x) = 49
    population std dev() = 5

Now we need to do hypothesis testing

* H0 = mean is still 50 
* Ha = mean is  50

* Let's take significance value () = 0.05

* Then we will calculate Z-statistic:

        (x-)/ (/sqrt(n)) = -1.26
* **This time this is two tailed test**, as in question it is written actual average weight differs significantly.
* We plot this -1.26 in the graph, also we will consider 1.26 as this is two tailed test.
* Now our p-value will be sum of both shaded areas. For left side area we can directly calculate using Z-table as it calculates area of LHS from a point. And we can multiply by 2 get total area( left+right) as it is identical.


* From Z-table we see a1 = 0.103, so total area (a1+a2) = 0.103 * 2 = 0.206
* So 0.206 is our p-value and this greater than significance value, hence we cannot reject null hypothesis.
