## Parametrics Tests:
### 1. T-Test:
* It was developed by William Gosset.
* **It is a small sample test. Size of sample here is < 30 (i.e n < 30)**.
* It is also called as student t test(probably it was some pen's name).

* x̄ is sample mean, µ is population mean, s is sample std deviation, n is sample' total population.
* **Here population variance is unknown.**
* Uses of T-test:
    * **A t-test is** a type of inferential statistic which is **used to determine if there is a significant difference between the means of two groups which may be related in certain features.**
* Assumption of T-test:
    * Sample size is small.
    * Population is normally distributed.
    * Samples are random and independent.

### ANOVA:
* It stands for Analysis of Variance.
* It was also developed by Fisher.
* It basically uses F-test to statistically to test equality of means and relative variance between them.


## Non-Parametric Tests:
### 1. Chi-Square Test:
* It was introduced by Karl Pearson.
* It is sampling analysis for testing significance of population variance.
* It uses simple random sampling method.
* It can be used for
    * Test of goodness of fit between set of observed values and expected values. Greater the  difference, greater the value of Chi-square. If no difference between observed and expected then Chi-square is 0.
    * As test of independence of two variables. For eg if we want to see relation between taste of music and gender, if they are related or not
* Chi-Square test values lies between 0 to 1.
* Conditions for Chi-Square test:
    * Sample size large(n>50)
    * Samples are independent


## When to use which Test:
* In hypothesis testing we come to a conclusion for a population based on sample dataset. Based on the sample dataset we divide into null(H0) and alternate hypothesis(HA), then we collect evidences and do different types of tests in order to find which hypothesis is true.
* Let's say we have a sample dataset: Age(M/F), Age_Group(Child, Adult, Elderly), Weight(continuous feature), Height(continuous feature).
* Remember in null hypothesis we write: An assumption that treates everything equal and similar. And opposite of null in alternate hypothesis.


### Case 1:

* Let's consider **one categorical feature** like Gender, then there are somethings which can be evaluated, one of the example is try to find whether there is a difference in proportion of male and female.
* Here:
    * H0: There is no difference in proportion
    * H1: There is difference in proportion
    * Then we take some significance value before test(let's say 0.05).
* Then as here we took only one categorical feature, then the test which we will apply is **One Sample Proportion Test**.
* And if this test gives value(p value)less than 0.05(significance value) we reject null hypothesis and accept alternate hypothesis.

### Case 2:
* Suppose we take **two categorical feature**, so questions can be is there any difference between male and female based on age_group.
* Then our null and alternate hypothesis will be same.
* Then here we apply **Chi-Square Test**..
* Similarly, if this test gives value less than 0.05(selected significance value) we reject null hypothesis and accept alternate hypothesis.

### Case 3:
* Suppose we take **one continuous numerical feature** let's say Height, so questions can be whether mean of population height and sample height of the sample are different or not.
* Here we use **T-Test**.


### Case 4:
* In case of **two numerical variable** we use **correlation**.
* Here null hypothesis can be there is a relation between two variables and alternate hypothesis can be there is no relation between two variables.

### Case 5:
* If we have **one numerical variable, one categorical variable and in categorical variable we have only two categories** then we use **T-Test**.
* If we have **one numerical variable and a categorical variable with more than two categories**, then we use **ANOVA Test**.
