## NLP Pipeline
* There are various steps like in the NLP pipeline:
    1. Data acquisition: When Data science team acquires data from some other team or tool
    2. Text extraction and cleanup: Taking only relevant texts or discarding text or fixing spelling mistakes, removing unnecesssary punctuations.
    3. Pre-processing
        * Sentence segmentation(split into separate sentences) and then word tokenization (splitting into words), then using stemming or lemmatization using nltk or spacy
    4. Feature engineering: Convert string words to numbers
    5. Model Building: Using machine learning algorithm
    6. Evaluation: We can use Grid Search CV to build perfect model and again perform same step 4.
        * Now we can check using confusion matrix, accuracy, recall, precision, F1-score(as it was classification task) how our model has performed on testing data.
    7. Deployment: Deploy it on server using Flask/any other and some cloud service.
    8. Monitor & Update


## Spacy vs NLTK
* Spacy is object oriented and NLTK is string processing library
* Spacy provides most efficient NLP algo for a given task. Hence if you care about end result, go with Spacy. NLTK provides access to many algos and if you care about specific algo and customizations and want to do some tweaks, then use NLTK.
* Spacy is new library an has active community, NLTK is old library but it also has active community.
