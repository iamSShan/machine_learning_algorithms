## Soft vs Hard Margin:
In hard margin there is no errors i.e point don't lie between margin and hyperplane. In soft margin we can have some errors, let' say our `c` value is 3, so 3 misclassified points can lie between hyperplane and margin.

### Kernels:


* Let's say we have **non linear dataset**(data which cannot be separated using single line or 2D plane), then we can't use same approach that we discussed in our intuition. We can't draw a straight line here. Here we will **use SVM kernel trick**.
* SVM algorithms use a set of mathematical functions that are defined as the kernel. The function of kernel is to take data as input and transform it into the required form. Different SVM algorithms use different types of kernel functions. These functions can be different types. For example linear, nonlinear, polynomial, radial basis function (RBF), and sigmoid.



* **In Kernel trick, the basic idea is that when a data set is inseparable in the current dimensions, add another dimension, maybe that way the data will be separable. Just think about it, the example above(in image) is in 2D and it is inseparable, but maybe in 3D there is a gap between the green and red points, maybe there is a level difference, so green points are on level one and red ones are on level two. In this case we can easily draw a separating hyperplane (in 3D a hyperplane is a plane) between level 1 and 2.**

* This tranformation from one dimension to higher dimension is done using some mathematical formulas. This transformation is called kernel transformation.
