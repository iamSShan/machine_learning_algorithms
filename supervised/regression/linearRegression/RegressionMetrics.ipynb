* For evaluation metrics like Mean Squared Error (MSE), Root Mean Squared Error (RMSE), and Mean Absolute Error (MAE) in regression analysis, we generally aim for lower values.
* While R-squared and adjusted R-squared we want higher values. 

## RMSE(Root Mean Squared Error):
* It is basically root of MSE
* It has all property same as MSE
* Main benefit of RMSE is that the output of RMSE is in same unit just as output variable(y).
* It also has same disadvantage like MSE that it is not robust to outliers.

## R-squared(coefficient of determination or goodness of fit):
* Let's say we have feature like cgpa and we are trying to predict package. For a new data point, a random solution without considering cgpa column the package output can be given by taking average number.
* But we have cgpa column, so we use linear regression and draw best fit line, to give better solutions.
* R-squared score is that linear regression line how much better than mean line.



* R² is used to check how good a specific model is
* R² = variance explained by model / total variance

* The formula of this is:
      1 - SSR/SStot
    
      where SSR is sum of squared error in regression line
      SStot is sum of squared error in mean line
      
      SSR and SStot will be written as:
      
      1 - Σ(yᵢ - ŷᵢ)² / Σ(yᵢ - ȳᵢ)²
      
      where ŷ is predicted y and ȳ is mean

* **How to interpret R² score:**
  * If R² is 0 when best fit line and mean line is same, model is not able to explain variance in target variable around its mean
  * For R² to be 1, this term Σ(yᵢ - ŷᵢ)² i.e. SSR has to be 0, which means best fit regression line is not doing any mistake i.e. line which is passing through each data point(which is not possible in real world data)
  
  * So we want R² score close to 1 to say it is a good model.
  
  * Let's say **if we got R² as 0.80** and let's say our independent feature was cgpa and output(dependent) feature is package then we can say that **cgpa(input column) is able to explain 80% variance in package**. Means 80% of different packages for each cgpa are known to model, except 20% which could be due to some external factors.
  
  
  * **What if R² is negative?** It is rare case but sometimes seen, here this ratio Σ(yᵢ - ŷᵢ)² / Σ(yᵢ - ȳᵢ)² has to be greater than 1 and it can be greater than 1 when Σ(yᵢ - ŷᵢ)²  > Σ(yᵢ - ȳᵢ)²  i.e. SSR > SStot, which means the best fit regression line is worst even more than mean line. 
    * Like if non-linear data if we apply linear regression, here we could get R² score negative.

## Adjusted R-squared:
* If we add features to input data, the R² increases, it is assumed that as more features we give, more variance it is able to explain. But if we add some irrelevant column as feature which is not correlated to output variable  , R² score will still increase or will remain same, when ideally is should have decreased.
* To handle this issue of R² score, adjusted R² score is used.

* Then in above formula let's talk about term on RHS of `1 -`
  * p will increase so (n-p-1) will decrease, in numerator (n-1) will remain constant and R² will increase a bit or will remain same as irrelevant feature was added, so (1-R²) will decrease or remain same, but denominator is decreasing in more mangnitude.
  * Numerator is almost remaining constant and denominator is decreasing so we can whole term is increasing
* Now 1 - increased value, this will give less adjusted R² score.


### Case 1: If we are adding a feature which is relevant and correlated with output variable

* In above formula let's talk about term on RHS of `1 -`
  * Simlar to above case, p will increase so (n-p-1) will decrease, in numerator (n-1) will remain constant.
  * But as we added relevant column then R² score will increase more, so (1-R²) decrease will be more than (n-p-1) decrease
  * So, whole numerator will decrease and whole term is decreasing
* Now 1 - decrease value, this will give more adjusted R² score.
