## Introduction:
**Binary classification**: It is when we have to classify objects into two groups. Generally, these two groups consist of ‘0’ and ‘1’. For e.g: determining whether a person has diabetes or not.

**Multi-class classification**:  Here there are more than two classes. For example, given a set of attributes of fruit, like it’s shape and colour, a multi-class classification task would be to determine the type of fruit.


One-vs-Rest (OVR) Method:
Many popular classification algorithms were designed natively for binary classification problems. These algorithms include :
 * Logistic Regression
 * Support Vector Machines (SVM)
 * Perceptron Models
 * Etc.
    
## Working:
* For example: there is a classification problem is to classify various fruits into three types of fruits: apple, guava or mango. Now, this is clearly a multi-class classification problem. Since there are three classes in the classification problem, the One-vs-Rest method will break down this problem into three binary classification problems:

        Problem 1 : apple vs [guava, mango]
        Problem 2 : guava vs [apple, mango]
        Problem 3 : mango vs [apple, guava]
* In problem 1 guava and mango are treated as same category and similarly for other problems.
* So instead of solving it as (apple vs guava vs mango), it is solved using three binary classification problems as shown above.
* Three models will be created in this example, 1st model considering Problem 1, and then 2nd model is created for Problem 2 and similarly 3rd model for Problem 3.
* For example: https://www.youtube.com/watch?v=V8fS0T_ktn4&t=142s
* For test data, when prediction is made, these 3 models gives us a probability values. Then we consider the either first or second or third target feature value based on highest probabilty.
* A major downside or disadvantage of this method is that many models have to be created. For a multi-class problem with ‘n’ number of classes, ‘n’ number of models have to be created, which may slow down the entire process. However, it is very useful with datasets having a small number of classes, where we want to use a model like SVM or Logistic Regression.
